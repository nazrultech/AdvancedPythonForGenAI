```python
df=pd.read_csv('D:\\IITM\\Advanced Python\\Dr Darshan Practice\\student_records.csv')
df
```

```python
# Add  a new column in the above
df["Total"] = df["ResearchScore"] + df["ProjectScore"]
df
```

```python
# Add percentage column

df["Percentage"] = (df["Total"]/200) * 100
df

# No need to use for loop, it comes automatilly by the use of Pandas. This is advanced Python :)
```

```python
# Round off the percentage column to 2 decimal places - round() or np.round()
# np.round() i fast compared to basic round function

import numpy as np
df["Percentage"] = np.round(df["Percentage"], 0)
df
```

```python
# Check datatype of each column - dtypes
df.dtypes

# in pndas strings are nothing but object
```

```python
df.info()
```

```python
df
```

```python
# Convert the data type of percentage column to int -astype()
# since float is taking more memory
df["Percentage"] =df["Percentage"].astype(int)
df
```

```python
# sort the students on the descending order of the percentage column - sort_values()

df= df.sort_values(by=["Percentage"], ascending =False)
df
```

```python
# in the above code indexes are changed
# resetting the index - reset_index()
# df =df.reset_index(drop=True)
df.reset_index()
```

```python
# here in the above code index is orgnixed but a new index column is created
df =df.reset_index(drop=True)
df
```

```python
# Fiiltering on the dataframe
# find all the names who have scored more than 80% -Filtering

df["Percentage"] >80
```

```python
# from the dataframe
df[df["Percentage"] >80]
```

```python
# Display only name
df[df["Percentage"] >80]["Name"]
```

```python
# dataframe

df
```

```python
# find all the names of students who have scored more than 70 in Researchscore and more than 80 in ProjectScore
# from the above answer is only Henry

df[(df["ResearchScore"] > 70) & (df["ProjectScore"] > 80)] 
```

```python
# if we need only name
df[(df["ResearchScore"] > 70) & (df["ProjectScore"] > 80)] ["Name"]
```

```python
df
```

```python
# create a coloumn "Status" nd assign "Pass" if the student has scored more than 40 in Total column else fail

def calculate_result(per):
    if per>40:
        return "Pass"
    else:
        return "Fail"
```

```python
df["Status"] = df["Percentage"].apply(calculate_result)
df
```

```python
df.head(2)
```

```python
# groupby()- It is used to split the data into groups based on some criteria
df.groupby(["Status"]).mean(numeric_only=True)["Percentage"]
```

```python
# df[Rows , COLs]
# iloc - Implicit location
# implicit locatio are the location of the rows and columns in the DataFrame which are not visible to us
# however we can use them using iloc

df.iloc[0:5, 0:3]
```

```python
# explicit location

df.loc[0:4 , "Name": "Obedient"]
```
